```
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
from sklearn.cluster import Birch
```

```
blobs = 5
X, y = make_blobs(n_samples=500, n_features=2, centers=blobs, random_state=2)

plt.scatter(X[:, 0], X[:, 1], marker="^", c=y)
plt.show()
```

```
model = KMeans(n_clusters=blobs)
model.fit(X)
pred = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], marker="^", c=pred)
plt.show()
```

```
model = Birch(n_clusters=blobs)
model.fit(X)
pred = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], marker="^", c=pred)
plt.show()
```

```
Xc, yc = make_circles(n_samples=500, random_state=2)
plt.scatter(Xc[:, 0], Xc[:, 1], marker='.', c = yc)
plt.show()
```

```
model = KMeans(n_clusters=8)
model.fit(Xc)
predc = model.predict(Xc)

plt.scatter(Xc[:, 0], Xc[:, 1], marker='.', c = predc)
plt.show()
```
